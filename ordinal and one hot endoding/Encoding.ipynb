# One Hot Encoding using Pandas

```python
df = pd.read_csv('cars.csv')
```

```python
df.head()
```

```python
df['owner'].value_counts()
```

```python
df['fuel'].value_counts()
```

```python
#we use func get_dummies and pass the df name and list of cols we want to use
#no of columns in new dataframe will be 2 less than the oroginal 

#We will get 4 new cols for fuel and 5 new columns for owner

pd.get_dummies(df,columns=['fuel','owner'])
```

```python
#ONE HOT ENCODING DOES NOT SOLVE THE PROBLEM OF MULTICOLLINEARITY 
```

# 2. K-1 OneHotEncoding

```python
#TO SOLVE MULTICOLLINEARITY, WE USE drop_first
pd.get_dummies(df,columns=['fuel','owner'],drop_first=True)
```

```python
#WE DO NOT USE PANDAS FOR ML PROJECT as it does not remember the position of columns
#We use pandas for one hot encoding only for the analysis
#We use one hot encoder class instead in sklearn

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.iloc[:,0:4],df.iloc[:,-1],test_size=0.2,random_state=2)
```

```python
X_train.head()
```

```python
X_test.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
#we want to apply one hot encoding only on fuel and owner
#we extract these 2 ols and then append to other cols > sol is column transform

ohe = OneHotEncoder(drop='first')
```

```python
X_train_new = ohe.fit_transform(X_train[['fuel', 'owner']])
```

```python
X_test_new = ohe.transform(X_test[['fuel', 'owner']])
```

```python
X_train_new.shape
```

```python
print(X_train_new.shape, X_test_new.shape)
```

```python
# Concatenate brand and km_driven columns with the encoded features (X_train_new)
X_train_concat = np.hstack((X_train[['brand', 'km_driven']].values, X_train_new.toarray()))

# Similarly, concatenate for X_test
X_test_concat = np.hstack((X_test[['brand', 'km_driven']].values, X_test_new.toarray()))

```
